```python
%matplotlib inline
import torch
import numpy as np
import matplotlib.pylab as plt
import sys
sys.path.append('..')
import d2lzh_pytorch as d2l

```

```python
def xyplot(x_vals, y_vals, name):
    d2l.set_figsize(figsize=(5, 2.5))
    d2l.plt.plot(x_vals.detach().numpy(), y_vals.detach().numpy())
    d2l.plt.xlabel('x')
    d2l.plt.ylabel(name + '(x)')
```

```python
x = torch.arange(-8.0, 8.0, 0.01, requires_grad=True)
```

```python
y = x.relu()
xyplot(x, y, 'relu')
```

```python
y.sum().backward()
xyplot(x, x.grad, 'grad of relu')
```

```python
y = x.sigmoid()
xyplot(x, y, 'sigmod')
```

```python
x.grad.zero_()
y.sum().backward()
xyplot(x, x.grad, 'grad of sigmod')
```

```python
y = x.tanh()
xyplot(x, y, 'tanh')
```

```python
x.grad.zero_()
y.sum().backward()
xyplot(x, x.grad, 'grad of tanh')
```
